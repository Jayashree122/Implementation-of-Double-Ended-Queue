```python
#Implementation of Double Ended Queue
class Deque:
    def __init__(self,size):
        self.q = [None]*size;
        self.f = -1
        self.r = -1
        self.size = size
    def isfull(self):
        return self.f == 0 and self.r == self.size-1
    def isempty(self):
        return self.f == -1 and self.r == -1
    def enque_at_front(self,ele):
        if self.isfull():
            raise Exception("Queue is full")
        elif self.isempty():
            self.f = self.r = 0
            self.q[self.f] = ele
        elif self.f == 0:
            self.f = self.size - 1
            self.q[self.f] = ele
        else:
            self.f -= 1
            self.q[self.f] = ele
    def enque_at_end(self,ele):
        if self.isempty():
            self.f = self.r = 0
            self.q[self.r] = ele
        elif self.isfull():
            raise Exception("Queue is full")
        elif self.r == self.size -1:
            self.r = 0
            self.q[self.r] = ele
        else:
            self.r += 1
            self.q[self.r] = ele
    def getfront(self):
        if self.isempty():
            raise Exception("Queue is empty")
        else:
            return self.q[self.f]
    def getrear(self):
        if self.isempty():
            raise Exception("Queue is empty")
        else:
            return self.q[self.r]
    def deq_at_front(self):
        if self.isempty():
            raise Exception("Queue is empty")
        elif self.f == self.r:
            self.f = self.r = -1
        elif self.f == self.size -1:
            self.f = 0
        else:
            self.f += 1
    def deq_at_rear(self):
        if self.isempty():
            raise Exception("Queue is full")
        elif self.r == self.f:
            self.r = self.f = -1
        elif self.r == 0:
            self.r = self.size - 1
        else:
            self.r -= 1
    def traverse(self):
        if self.isempty():
            raise Exception("Queue is empty")
        i = self.f
        while i!=self.r:
            print(self.q[i],end = " ")
            i = (i+1)%self.size
        print(self.q[i],end = " ")
```

```python
c=Deque(5)
c.enque_at_front(33)
c.enque_at_front(22)
c.deq_at_front()
c.enque_at_front(11)
c.enque_at_end(44)
c.enque_at_end(55)
c.traverse()
c.deq_at_rear()
```
